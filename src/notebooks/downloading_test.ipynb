```python
import os
import io
import math
import itertools
import concurrent.futures
from typing import Any, Dict, List, Tuple
from osgeo import gdal, osr
from box_cls import Box, BoxInt, box_pixels_to_coordinates

import numpy as np
from PIL import Image

import httpx


gdal.UseExceptions()

# Start of the function


def download_cir(
    image_coords_box: Box, resolution: float, save_path: str, quiet: bool = False, **kwargs
):
    url = "https://services.arcgisonline.nl/arcgis/rest/services/Luchtfoto/Luchtfoto_CIR/MapServer/export"
    SESSION = httpx.Client()

    web_mercator = osr.SpatialReference()
    web_mercator.ImportFromEPSG(28992)

    WKT_28992 = web_mercator.ExportToWkt()
    BASE_MAP_SCALE = 3779.52

    download_resolution = 0.25
    coord_size_x = image_coords_box.x_max - image_coords_box.x_min
    coord_size_y = image_coords_box.y_max - image_coords_box.y_min
    total_pixels_x = coord_size_x / download_resolution
    total_pixels_y = coord_size_y / download_resolution
    pixels_per_download = 380  # Pixel size of each individual image that is downloaded
    coords_per_download = (
        pixels_per_download * download_resolution
    )  # Coordinates size of each individual image that is downloaded

    image_pixels_box = BoxInt(
        x_min=0, y_min=0, x_max=math.ceil(total_pixels_x), y_max=math.ceil(total_pixels_y)
    )

    map_scale = BASE_MAP_SCALE * download_resolution

    def get_tile(url: str, params: Dict[Any, Any]):
        retry = 3
        while 1:
            try:
                r = SESSION.get(url, params=params, timeout=60)
                break
            except Exception:
                retry -= 1
                if not retry:
                    raise
        if r.status_code == 404:
            return None
        elif not r.content:
            return None
        r.raise_for_status()
        return r.content

    def is_empty(im: Image.Image):
        extrema = im.getextrema()
        if len(extrema) >= 3:
            if len(extrema) > 3 and extrema[-1] == (0, 0):
                return True
            for ext in extrema[:3]:
                if ext != (0, 0):
                    return False
            return True
        else:
            return extrema[0] == (0, 0)  # type: ignore

    def paste_tile(
        bigim: Image.Image | None, tile: bytes | None, pixels_box: BoxInt, image_pixels_box: BoxInt
    ):
        if tile is None:
            return bigim
        im = Image.open(io.BytesIO(tile))
        mode = "RGB" if im.mode == "RGB" else "RGBA"
        print(f"{mode = }")
        if bigim is None:
            newim = Image.new(
                mode,
                (
                    image_pixels_box.x_max - image_pixels_box.x_min,
                    image_pixels_box.y_max - image_pixels_box.y_min,
                ),
            )
        else:
            newim = bigim

        xy0 = (pixels_box.x_min, pixels_box.y_min)
        if mode == "RGB":
            newim.paste(im, xy0)
        else:
            if im.mode != mode:
                im = im.convert(mode)
            if not is_empty(im):
                newim.paste(im, xy0)
        im.close()
        return newim

    def finish_picture(bigim: Image.Image, image_pixels_box: BoxInt):
        crop = (
            image_pixels_box.x_min,
            image_pixels_box.y_min,
            image_pixels_box.x_max,
            image_pixels_box.y_max,
        )
        retim = bigim.crop(crop)
        if retim.mode == "RGBA" and retim.getextrema()[3] == (255, 255):  # type: ignore
            retim = retim.convert("RGB")
        bigim.close()
        return retim

    print(f"{image_coords_box = }")
    print(f"{image_pixels_box = }")
    print(f"{coords_per_download = }")

    iter_pixels_corners = list(
        itertools.product(
            range(image_pixels_box.x_min, image_pixels_box.x_max, pixels_per_download),
            range(image_pixels_box.y_min, image_pixels_box.y_max, pixels_per_download),
        )
    )

    print(f"{iter_pixels_corners = }")

    iter_pixels_corners_all = list(
        map(
            lambda t: (t[0], t[1], t[0] + pixels_per_download, t[1] + pixels_per_download),
            iter_pixels_corners,
        )
    )

    print(f"{iter_pixels_corners_all = }")

    iter_pixels_boxes = list(map(BoxInt, *zip(*iter_pixels_corners_all)))

    print(f"{iter_pixels_boxes = }")

    totalnum = len(iter_pixels_boxes)
    futures: List[concurrent.futures.Future[bytes | None]] = []
    with concurrent.futures.ThreadPoolExecutor(5) as executor:
        for pixels_box in iter_pixels_boxes:
            coords_box = box_pixels_to_coordinates(
                pixels_box,
                image_pixels_box=image_pixels_box,
                image_coordinates_box=image_coords_box,
            )
            current_bbox = (
                f"{coords_box.x_min},{coords_box.y_min},{coords_box.x_max},{coords_box.y_max}"
            )
            params = {
                "f": "image",  # Output format
                "bboxSR": "EPSG:28992",  # Coordinate system
                "bbox": current_bbox,
                "size": f"{pixels_per_download},{pixels_per_download}",  # Width and height of the output image
                "mapScale": map_scale,
                "formatOptions": "TIFF",
                "transparent": True,
            }
            futures.append(executor.submit(get_tile, url, params))
        bigim = None
        for k, (fut, pixels_box) in enumerate(zip(futures, iter_pixels_boxes), 1):
            bigim = paste_tile(bigim, fut.result(), pixels_box, image_pixels_box)
            if not quiet:
                print(f"Downloaded image {str(k).zfill(len(str(totalnum)))}/{totalnum}")
    if bigim is None:
        print("Nothing was found at these coordinates.")
        return

    if not quiet:
        print("Saving GeoTIFF. Please wait...")
    img = finish_picture(bigim, image_pixels_box)
    print(f"{type(img) = }")

    real_total_pixels_x = math.ceil(coord_size_x / resolution)
    real_total_pixels_y = math.ceil(coord_size_y / resolution)

    img = img.resize((real_total_pixels_x, real_total_pixels_y))

    imgbands = len(img.getbands())
    driver = gdal.GetDriverByName("GTiff")

    if "options" not in kwargs:
        kwargs["options"] = [
            "COMPRESS=DEFLATE",
            "PREDICTOR=2",
            "ZLEVEL=9",
            "TILED=YES",
        ]

    try:
        gtiff = driver.Create(
            save_path,
            img.size[0],
            img.size[1],
            imgbands,
            gdal.GDT_Byte,
            **kwargs,
        )
        gtiff.SetGeoTransform(
            (image_coords_box.x_min, resolution, 0, image_coords_box.y_max, 0, -resolution)
        )
        gtiff.SetProjection(WKT_28992)
        for band in range(imgbands):
            array = np.array(img.getdata(band), dtype="u8")
            array = array.reshape((img.size[1], img.size[0]))
            band = gtiff.GetRasterBand(band + 1)
            band.WriteArray(array)
        gtiff.FlushCache()
    except Exception as e:
        raise e
    finally:
        gtiff = None

    if not quiet:
        print(f"Image saved to {save_path}")


save_path = "New_image_7.tif"

coord_size = 100
base_bbox = (122000, 483000)
resolution = 0.08
kwargs = {}
quiet = False

image_coords_box = Box(
    x_min=base_bbox[0],
    y_min=base_bbox[1],
    x_max=base_bbox[0] + coord_size,
    y_max=base_bbox[1] + coord_size,
)

download_cir(image_coords_box=image_coords_box, resolution=resolution, save_path=save_path)
```
